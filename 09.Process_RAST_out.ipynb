```r
library(dplyr)
library(tidyverse)
library(purrr)
```

```r
## Load the two TSV files into R (the category table and the feature table for each isolate)
features_SA1 <- read_tsv("data/features_SA1.tsv")  
features_SA6 <- read_tsv("data/features_SA6.tsv")
features_SB1 <- read_tsv("data/features_SB1.tsv")
features_SB8 <- read_tsv("data/features_SB8.tsv")
features_WA5 <- read_tsv("data/features_WA5.tsv")
features_WB3 <- read_tsv("data/features_WB3.tsv")
features_WB8 <- read_tsv("data/features_WB8.tsv")
SA1_df <- read_tsv("data/SA1_df.tsv")
SA6_df <- read_tsv("data/SA6_df.tsv")
SB1_df <- read_tsv("data/SB1_df.tsv")
SB8_df <- read_tsv("data/SB8_df.tsv")
WA5_df <- read_tsv("data/WA5_df.tsv")
WB3_df <- read_tsv("data/WB3_df.tsv")
WB8_df <- read_tsv("data/WB8_df.tsv")
```

```r
head(features_SA1)
head(SA1_df)
```

```r
## Remove NA columns 
clean_na_columns <- function(df) {
  df %>%
    select_if(~ !all(is.na(.)))
} # function to remove NA columns

features_SA1 <- clean_na_columns(features_SA1)
features_SA6 <- clean_na_columns(features_SA6)
features_SB1 <- clean_na_columns(features_SB1)
features_SB8 <- clean_na_columns(features_SB8)
features_WA5 <- clean_na_columns(features_WA5)
features_WB3 <- clean_na_columns(features_WB3)
features_WB8 <- clean_na_columns(features_WB8)
SA1_df <- clean_na_columns(SA1_df)
SA6_df <- clean_na_columns(SA6_df)
SB1_df <- clean_na_columns(SB1_df)
SB8_df <- clean_na_columns(SB8_df)
WA5_df <- clean_na_columns(WA5_df)
WB3_df <- clean_na_columns(WB3_df)
WB8_df <- clean_na_columns(WB8_df) # apply the function to each dataframe 
```

```r
head(SA1_df)
head(features_SA1)
```

```r
## Make 'Features' column in long format, separating each feature identified using commas
make_long_format <- function(df) {
  df %>%
    separate_rows(Features, sep = ",")
} # Function to make 'Features' column in long format

SA1_df <- make_long_format(SA1_df)
SA6_df <- make_long_format(SA6_df)
SB1_df <- make_long_format(SB1_df)
SB8_df <- make_long_format(SB8_df)
WA5_df <- make_long_format(WA5_df)
WB3_df <- make_long_format(WB3_df)
WB8_df <- make_long_format(WB8_df)
```

```r
# Rename features column name on features_dfs so it has the same name on both dataframes 
## Define the renaming function:
rename_features <- function(df) {
  df %>%
    rename(Features = `Feature ID`)
}

## Apply the function to all relevant data frames
df_names <- ls(envir = .GlobalEnv, pattern = "^features_") # get df names 

df_names %>%
  map(~ {
    df <- get(.x, envir = .GlobalEnv)
    df <- rename_features(df)
    assign(.x, df, envir = .GlobalEnv)
  }) # Retrieve and rename all matching data frames

head(features_SA1)
```

```r
## Merge both dataframes in rows where there's a match between features in both dataframes, filtering out columns I'm not interested in 
columns_to_keep <- c("Category", "Subsystem", "Role",  "Contig", "Start", "Stop")  # Define the columns to keep

merged_SA1 <- inner_join(SA1_df, features_SA1, by = "Features") %>% select(all_of(columns_to_keep))
merged_SA6 <- inner_join(SA6_df, features_SA6, by = "Features") %>% select(all_of(columns_to_keep))
merged_SB1 <- inner_join(SB1_df, features_SB1, by = "Features") %>% select(all_of(columns_to_keep))
merged_SB8 <- inner_join(SB8_df, features_SB8, by = "Features") %>% select(all_of(columns_to_keep))
merged_WA5 <- inner_join(WA5_df, features_WA5, by = "Features") %>% select(all_of(columns_to_keep))
merged_WB3 <- inner_join(WB3_df, features_WB3, by = "Features") %>% select(all_of(columns_to_keep))
merged_WB8 <- inner_join(WB8_df, features_WB8, by = "Features") %>% select(all_of(columns_to_keep))

head(merged_SA1)
```

```r
## Save them as tsv files 
write_tsv(merged_SA1, "genome_mining-CARB_SA1.tsv")
write_tsv(merged_SA6, "genome_mining-CARB_SA6.tsv")
write_tsv(merged_SB1, "genome_mining-CARB_SB1.tsv")
write_tsv(merged_SB8, "genome_mining-CARB_SB8.tsv")
write_tsv(merged_WA5, "genome_mining-CARB_WA5.tsv")
write_tsv(merged_WB3, "genome_mining-CARB_WB3.tsv")
write_tsv(merged_WB8, "genome_mining-CARB_WB8.tsv")
```

```r
# Combine all files into a single one for future processing and plotting 
## set wd in data to ease 
setwd("/home/joao/Documents/Research/Collaborations/Mangrove_Microbiota/WGS/genome-mining/data")

## Get all files 
file_pattern <- "genome_mining-CARB_" # Define the pattern for files to read
file_list <- list.files(pattern = file_pattern) # List all files matching the pattern
file_list
```

```r
## Function to read each file and add the Isolate column
read_and_label <- function(file) {
  df <- read_tsv(file)
  isolate_name <- gsub("genome_mining-CARB_|\\.tsv", "", basename(file))
  df %>%
    mutate(Isolate = isolate_name)
}

## Read, process, and combine all files
combined_df <- file_list %>%
  lapply(read_and_label) %>%  # Read each file and add Isolate column
  bind_rows()                 # Combine all data frames into one

## Arrange columns with 'Isolate' first
combined_df <- combined_df %>%
  select(Isolate, everything())

head(combined_df)
write_tsv(combined_df, "RAST_CARB_rawoutput.txt")
```
